```python
plt.figure(figsize=(16 , 6))
property_types = df['room type'].value_counts().to_frame()
room_type_bar = plt.bar(property_types.index, property_types.loc[:,"count"] )
plt.bar_label(room_type_bar, labels = property_types.loc[:,"count"], padding = 4 )
plt.ylim([0,50000])
plt.xlabel('Room Type')
plt.ylabel('Room Type Count')
plt.title('Property Types and their count in the Dataset')
```

```python
hood_group = df['neighbourhood group'].value_counts().to_frame()
hood_group
```

```python
plt.figure(figsize=(16 , 6))

hood_group_bar = plt.bar(hood_group.index, hood_group.loc[:,"count"])
plt.bar_label(hood_group_bar, labels = hood_group.loc[:,"count"], padding = 4 )
plt.ylim([0,40000])
plt.xlabel('Neighborhood Groups')
plt.ylabel('Number of Listings')
plt.xticks(rotation = 45)
plt.title('Which Neighborhood Groups has the highest number of Listings')
```

```python
plt.figure(figsize=(16 , 6))
avg_price = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending = False).to_frame()
avg_price_bar = plt.bar(avg_price.index, avg_price.loc[:,'price'])
plt.bar_label(avg_price_bar, labels = round(avg_price.loc[:,"price"], 2), label_type = 'edge', padding = 4 )
plt.ylim([0,700])
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price per Listings ($)')
plt.xticks(rotation = 45)
plt.title('Average Price per Listings ($) in each Neighborhood Group')
```

```python
plt.figure(figsize=(16 , 6))

df.groupby(df['Construction year'])['price'].mean().to_frame().plot()
plt.xlabel('Construction Year')
plt.ylabel('Average Price ($)')
plt.title('Average Price ($) for Properties in each Construction Year')
plt.show()
```

```python
plt.figure(figsize=(16 , 6))

hosts = df.groupby('host name')['calculated host listings count'].sum().sort_values(ascending = False).nlargest(10).to_frame()
hosts_bar = plt.bar(hosts.index, hosts.loc[:, 'calculated host listings count' ])
plt.bar_label(hosts_bar, label = hosts.loc[:, 'calculated host listings count'], label_type = 'edge', padding = 3)
plt.xlabel('Hosts Name')
plt.ylabel('Calculated Host Listings Count')
plt.xticks(rotation = 45)
plt.ylim([0, 120000])
plt.title('Top 10 Hosts by Calculated host Listings Count')
```

# **Business Insights**
--- 

## **What is the total number of listings ?**

```python
total_listings = df['id'].nunique()
print(f"Total Number of Listings: {total_listings}")
```

## **What is the average price of listings ?**

```python
average_price = df['price'].mean()
print(f"Average Price of Listings: ${average_price:.2f}")
```

## **What is the average number of reviews per listing ?**

```python
average_reviews = df['number of reviews'].mean()
print(f"Average Number of Reviews per Listing: {average_reviews:.2f}")
```

## **What is the proportion of listing with verified host identity ?**

```python
verified_host_proportion = df[df['host_identity_verified'] == 't'].shape[0] / df.shape[0]
print(f"Proportion of Listings with Verified Host Identity: {verified_host_proportion:.2%}")
```

## **What is the average service fee ?**

```python
average_service_fee = df['service fee'].mean()
print(f"Average Service Fee: ${average_service_fee:.2f}")
```

## **What is the avreage minimum nights required ?**

```python
average_minimum_nights = df['minimum nights'].mean()
print(f"Average Minimum Nights Required: {average_minimum_nights:.2f}")
```

## **Who are the top 5 hosts by number of listings ?**

```python
top_hosts = df['host name'].value_counts().sort_values(ascending= False).head(5)
print(f"Top 5 Hosts by Number of Listings:\n{top_hosts}")
```

## **What is the most common room type ?**

```python
most_common_room_type = df['room type'].mode()[0]
print(f"Most Common Room Type: {most_common_room_type}")
```

## **What is the average price by room type ?**

```python
average_price_by_room_type = df.groupby('room type')['price'].mean()
print(f"Average Price by Room Type:\n{average_price_by_room_type}")
```

## **What is the distribustion of listings by neighbourhood Group ?**

```python
listings_by_neighbourhood_group = df['neighbourhood group'].value_counts()
print(f"Distribution of Listings by Neighbourhood Group:\n{listings_by_neighbourhood_group}")
```

## **What is the average construction year of listings ?**

```python
average_construction_year = df['Construction year'].mean()
print(f"Average Construction Year of Listings: {average_construction_year:.0f}")
```

## **What is the canellation policy proportion ?**

```python
cancellation_policy_proportion = df['cancellation_policy'].value_counts(normalize=True)
print(f"Proportion of Listings by Cancellation Policy Type:\n{cancellation_policy_proportion}")
```
